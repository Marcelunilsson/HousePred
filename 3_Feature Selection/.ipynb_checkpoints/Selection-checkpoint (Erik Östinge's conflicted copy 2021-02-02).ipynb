```python
types = np.array(df["Objekttyp:S:O"].unique())
excludeType = np.array(['Radhus', 'Villa', 'Parhus', 'Fritidshus', 'Kedjehus',
       'Tomt/Mark', 'Hus', 'G√•rd', '√ñvrigt'])
df = df[df["Objekttyp:S:O"].isin([otype for otype in types if otype not in excludeType])]
```

```python
def inter_quartile_range_split(df):
    q1 = df['Pris:A(kr):O'].quantile(0.25)
    q3 = df['Pris:A(kr):O'].quantile(0.75)
    iqr = q3 - q1
    fence_low = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr
    df = df.loc[(df['Pris:A(kr):O'] > fence_low) & (df['Pris:A(kr):O'] < fence_high)]
    return df

df = inter_quartile_range_split(df)
```
